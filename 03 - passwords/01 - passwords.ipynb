```python
# Run this cell first
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
from IPython.display import HTML, display
from password_strength import PasswordStats
```

## Alapvető statisztikai jellemzők / basic statistic properties

### alphabet

a felhasznált karakterek halmaza / set of used characters

### alphabet_cardinality

felhasznált karakterek száma / alphabet length

### char_categories

Karakterek száma kategóriánként / Character count per category
- L: betű / letter
- M: jel / mark
- N: szám / number
- P: írásjel / punctuation
- S: szimbólum / symbol
- Z: elválasztó / separator
- C: egyéb / other

### length

hossz

### combinations

az aktuális betűk kombinációs lehetőségeinek száma / the number of possible combinations with the current alphabet

### entropy bits

a lehetséges jelszavak 2 alapú logaritmusa / log2 of the number of possible passwords

### strength

jelszó erősség / password strength

 * 0.00 - 0.33 gyenge / weak
 * 0.33 - 0.66 közepes / medium
 * 0.66 - 1.00 erős / strong

### repeated patterns length

az ismétlődő mintázatok hossza / length of repeated patterns

### sequences_length

betú vagy szám sorozatok hossza / length of used letter or numeric sequences

### weakness_factor

az ismétlődő vagy sorozatok aránya / portion of the password that contains patterns or sequences

```python
w3 = widgets.HTML(value="")
def password_stats(password = "secret1234aaaa"):
    if len(password) > 0:
        ps = PasswordStats(password)
        w3.value = "<table>"
        w3.value += "<tr><td>alphabet:</td><td>"+str(ps.alphabet)+"</td></tr>"
        w3.value += "<tr><td>alphabet cardinality:</td><td>"+str(ps.alphabet_cardinality)+"</td></tr>"
        w3.value += "<tr><td>character categories:</td><td>"+str(ps.char_categories)+"</td></tr>"
        w3.value += "<tr><td>length:</td><td>"+str(ps.length)+"</td></tr>"
        w3.value += "<tr><td>combinations:</td><td>"+str(ps.combinations)+"</td></tr>"
        w3.value += "<tr><td>repeated_patterns_length:</td><td>"+str(ps.repeated_patterns_length)+"</td></tr>"
        w3.value += "<tr><td>sequences_length:</td><td>"+str(ps.sequences_length)+"</td></tr>"
        w3.value += "<tr><td>weakness_factor:</td><td>"+str(ps.weakness_factor)+"</td></tr>"
        w3.value += "<tr><td>entropy bits:</td><td>"+str(ps.entropy_bits)+"</td></tr>"
        w3.value += "<tr><td>strength:</td><td>"+str(ps.strength())+"</td></tr>"
        w3.value += "<table>"
interact(password_stats)
display(w3)
```
